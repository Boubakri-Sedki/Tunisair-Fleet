# Introduction 

Tunisair serves 54 destinations in scheduled flights between Africa, Europe and the Middle East.<br>
The bulk of Tunisair's business is concentrated in Europe, particularly in France, with 9 routes (Paris Orly, Marseille, Nantes, etc.) compared with 5 in Italy and Germany (respectively Rome, Milan Venice and Frankfurt, Hamburg, Berlin). ..).<br>
The rest of the 32 European destinations are spread over the rest of the mainland countries such as Switzerland, London, Belgium, etc.<br>
Although belonging to the African continent, Tunisair serves only 14 African destinations for the majority concentrated on the north and west of Africa (Algiers, Casa, Tripoli, Benghazi, Oran, Dakar, Bamako, etc.). <br>
This choice was justified by socio-economic factors relating to the rest of the continent (low income, lack of appropriate infrastructure, political instability, etc.). <br>
However, the situation has changed and the indicators of the various components of the environment are improving, albeit slowly, but still continuing.<br>
Note in this regard, that in its 2012 recovery plan, the company plans the closure of unprofitable lines and the development of a set of services on the African continent benefiting from the growth of passenger and cargo traffic on the latter<br>
Like any company, Tunisair airline has in its assets, some that can be translated into competitive advantages but also handicaps that could negatively affect its development prospects in general and the African continent in particular.<br>
An internal diagnosis is able to list those elements that, faced with opportunities and threats, make it possible to set up more effective strategic actions.<br>

```python
from math import pi

cl0=delta[delta.cluster==0]
cl1=delta[delta.cluster==1]
cl2=delta[delta.cluster==2]

df = pd.DataFrame({
'cl': ["0","1","2"],
'class business': [cl0['Seat Width (Business)'].median(),cl1['Seat Width (Business)'].median(),cl2['Seat Width (Business)'].median()],
'class economy': [cl0['Seat Width (Economy)'].median(),cl1['Seat Width (Economy)'].median(),cl2['Seat Width (Economy)'].median()],
'first class': [cl0['Seat Width (First Class)'].median(),cl1['Seat Width (First Class)'].median(),
                cl2['Seat Width (First Class)'].median()],
'class eco comfort': [cl0['Seat Width (Eco Comfort)'].median(),cl1['Seat Width (Eco Comfort)'].median(),cl2['Seat Width (Eco Comfort)'].median() ]

})
#df.columns=['group','economy','business','first class','eco comfort']
 

# Set data

 
# ------- PART 1: Define a function that do a plot for one line of the dataset!
 
def make_spider( row, title, color):
    
    categories=list(df)[1:]
    N = len(categories)
 
    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
 
# Initialise the spider plot
    ax = plt.subplot(2,2,row+1, polar=True, )
 
# If you want the first axis to be on top:
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)
 
# Draw one axe per variable + add labels labels yet
    plt.xticks(angles[:-1], categories, color='grey', size=8)
 
# Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([10,20,30], ["10","20","30"], color="grey", size=8)
    plt.ylim(0,30)
 
# Ind1
    values=df.loc[row].drop('cl').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=2, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.4)

 
# Add a title
    plt.title(title, size=11, color=color, y=1.1)
 
# ------- PART 2: Apply to all individuals
# initialize the figure
my_dpi=96
plt.figure(figsize=(1000/my_dpi, 1000/my_dpi), dpi=my_dpi)
 
# Create a color palette:
my_palette = plt.cm.get_cmap("Set2", len(df.index))
 
# Loop to plot
for row in range(0, len(df.index)):
    make_spider( row=row, title='group '+str(df['cl'][row]), color=my_palette(row))

```
